```python
import turtle as t
pen=t.Turtle()
t.bgcolor("black")
pen.speed(200)
pen.pensize(4)
pen.pencolor("cyan")
def draw_Rectangle(a,b,col):
    pen.fillcolor(col)
    pen.begin_fill()
    for x in range(2):
        pen.forward(a)
        pen.left(90)
        pen.forward(b)
        pen.left(90)
    pen.end_fill()
def draw_triangle_90(a,b,col):
    pen.fillcolor(col)
    pen.begin_fill()
    pen.forward(a)
    pen.left(135)
    pen.forward(b)
    pen.left(90)
    pen.forward(b)
    pen.left(135)
    pen.forward(b)
    pen.end_fill()
def move(x,y):
    pen.penup()
    pen.goto(x,y)
    pen.pendown()
def draw_tree(a,b,c,col):
    pen.fillcolor (col)
    pen.begin_fill()
    pen.forward(a)
    pen.left(95)
    pen.forward(b)
    pen.left(85)
    pen.forward(c)
    pen.left(85)
    pen.forward(b)
    pen.left(95)
    pen.end_fill()
    

move(-150,0)
draw_Rectangle(200,100,"blue")
move(-65,100)
draw_Rectangle(30,50,"orange")
move(-150,100)
draw_triangle_90(125,90,"red")
move(-75,0)
draw_Rectangle(50,75,"yellow")
move(180,-50)

draw_tree(40,90,25,"gray")
move(215,40)
draw_triangle_90(60,110,"green")
move(220,120)
draw_triangle_90(40,90,"olive")
move(225,185)
draw_triangle_90(25,70,"lime")
move(-250,-90)

draw_tree(25,70,15,"gray")
move(-215,-20)
draw_triangle_90(40,90,"green")
move(-210,45)
draw_triangle_90(30,80,"olive")
move(-210,100)
draw_triangle_90(25,70,"lime")
move(-300,-300)
```
